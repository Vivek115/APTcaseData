Sort Stores in order of Efficacy

```python
if 'total Sales' not in dailyData17.columns:
    dailyData17.insert(len(dailyData17.columns),"total Sales", dailyData17.sum(axis=1))
    
if 'total Sales' not in dailyData18.columns:
    dailyData18.insert(len(dailyData18.columns),"total Sales", dailyData18.sum(axis=1))
    
ascendingStores17 = dailyData17.sort_values(by='total Sales').index
ascendingStores18 = dailyData18.sort_values(by='total Sales').index


print(ascendingStores17)
print(ascendingStores18)
```
