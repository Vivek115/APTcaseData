Aggregate total sales for Week

```python
dailyData17 = sevenAMToSixPM.iloc[:483,14:66] + sixPMToSevenPM.iloc[:483,14:66] + sevenPMToEightPM.iloc[:483,14:66] + eightPMToNinePM.iloc[:483,14:66]
dailyData18 = sevenAMToSixPM.iloc[:483,66:125] + sixPMToSevenPM.iloc[:483,66:125] + sevenPMToEightPM.iloc[:483,66:125] + eightPMToNinePM.iloc[:483,66:125]

dailyData17.insert(0, 'StoreID',sevenAMToSixPM['StoreID'])
dailyData18.insert(0, 'StoreID',dailyData17['StoreID'])
```

```python
dailyData17.insert(len(dailyData17.columns), 'State', stateData['State'])
dailyData18.insert(len(dailyData17.columns), 'State', stateData['State'])
```

```python
dailyData17 = dailyData17.set_index((dailyData17.columns[0]))
dailyData18 = dailyData18.set_index(dailyData18.columns[0])
```
